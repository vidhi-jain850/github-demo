```python
def new_decorate(original):
    def wrap():
        print("some extra code")
        
        original()
        print("some extra code")
    return wrap
```

```python
def fun_deco():
    print(" I want to be decorated")
```

```python
fun_deco()
```

```python
new_decorate(fun_deco)
```

```python
re=new_decorate(fun_deco)
```

```python
re
```

```python
re()
```

```python
@new_decorate
def fun_deco():
    print(" I want to be decorated")
```

```python
fun_deco()
```

```python
def cube(n):
    result=[]
    for i in  range(n):
        yield i**3
```

```python
cube(10)
```

```python
for i in cube(10):
    print(i)
```

```python
list(cube(10))
```

```python
def fibo(n):
    a=1
    b=1
    for i in range(n):
        yield a
        a,b=b,a+b
```

```python
list(fibo(10))
```
